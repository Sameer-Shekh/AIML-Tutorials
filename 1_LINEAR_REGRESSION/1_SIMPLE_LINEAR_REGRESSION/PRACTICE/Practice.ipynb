```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
canada = pd.read_csv('canada_per_capita_income.csv')
canada.head()
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada.year,canada[['per capita income (US$)']],marker='*',color='green')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(canada[['year']],canada['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada.year,canada['per capita income (US$)'],marker='*',color='purple')
plt.plot(canada.year,reg.predict(canada[['year']]),color='blue')
```

```python
reg.score(canada[['year']],canada['per capita income (US$)'])
```
