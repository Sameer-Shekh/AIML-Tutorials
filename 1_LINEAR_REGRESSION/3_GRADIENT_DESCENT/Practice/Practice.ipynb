```python
import numpy as np
import math

# Function for gradient descent
def gradient_descent(x, y):
    m_curr = b_curr = 0  # Starting values of m and b in y = mx + b
    iterations = 10000
    n = len(x)  # Assuming x and y have the same length
    learning_rate = 0.08  # Learning rate

    prev_cost = float('inf')  # Initialize previous cost to a high value

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr  # Predicted values (y = mx + b)
        
        # Compute gradients
        md = -(2/n) * sum(x * (y - y_predicted))  # Partial derivative w.r.t m
        bd = -(2/n) * sum(y - y_predicted)  # Partial derivative w.r.t b

        # Compute cost (Mean Squared Error - MSE)
        cost = (1/n) * sum((y - y_predicted) ** 2)

        # Check for convergence
        if math.isclose(cost, prev_cost, rel_tol=1e-20):
            break
        prev_cost = cost

        # Update m and b using gradient descent formulas
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        print(f"Iteration {i}: m={m_curr:.4f}, b={b_curr:.4f}, Cost={cost:.6f}")

# Example dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Run gradient descent
gradient_descent(x, y)
```
