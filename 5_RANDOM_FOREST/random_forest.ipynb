```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load the Digits dataset
digits = load_digits()

# Shape of data
print("Shape of digits.data:", digits.data.shape)  # (1797, 64)
print("Shape of digits.images:", digits.images.shape)  # (1797, 8, 8)

# Display the first digit using digits.images
plt.imshow(digits.images[0], cmap='gray')
plt.title(f"Digit: {digits.target[0]}")
plt.show()

# First row in digits.data (Flattened 8x8 image)
print("Flattened Image Data:", digits.data[0])
```

```python
import pandas as pd 
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df['target'] = digits.target
```

```python
df.head(15)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),df.target,test_size=0.2)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)*100
```

```python
#  THIS IS THE RESULT THAT IS GENERATED BY THE MODEL
model.predict(X_test)
```

```python
# THIS IS THE ACTUAL DATA CORRESPONDING TO THE X_TEST (BOTH THE DATAS HAVE SAME VALUES)
y_test
```

```python
import seaborn as sn
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred=y_predicted)
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
